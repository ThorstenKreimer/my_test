```python
import pandas as pd
import matplotlib.pyplot as plt

# Lade den Datensatz
df = pd.read_csv("wetter.csv")

# Konvertiere die 'Datum'-Spalte in ein datetime-Format
df['Datum'] = pd.to_datetime(df['Datum'])

# Extrahiere den Monat und das Jahr
df['Monat'] = df['Datum'].dt.month
df['Jahr'] = df['Datum'].dt.year

# Berechne die durchschnittliche Temperatur je Monat
# Gruppiere nach Jahr und Monat und berechne den Mittelwert der Temperatur
monatsmittel = df.groupby(['Jahr', 'Monat'])['Temperatur'].mean().reset_index()

# Erstelle das Balkendiagramm
plt.figure(figsize=(10, 6))
plt.bar(monatsmittel['Monat'], monatsmittel['Temperatur'])
plt.xlabel('Monat')
plt.ylabel('Durchschnittstemperatur (C)')
plt.title('Durchschnittstemperatur je Monat')
plt.xticks(range(1, 13))  # Monate von 1 bis 12
plt.show()

import pandas as pd
import matplotlib.pyplot as plt

# Lade den Datensatz
df = pd.read_csv("wetter.csv")

# Entferne Zeilen mit fehlenden Temperaturwerten, um das Histogramm korrekt darzustellen
df = df.dropna(subset=['Temperatur'])

# Erstelle das Histogramm fr die Temperaturverteilung
plt.figure(figsize=(10, 6))
plt.hist(df['Temperatur'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Temperatur (C)')
plt.ylabel('Hufigkeit')
plt.title('Verteilung der Temperaturen')
plt.show()
```
